```python
import numpy as np
import matplotlib.pyplot as plt

user_input = input("Enter Activation Function to plot (eg. linear, relu, sigmoid, tanh): ").lower().split()

x = np.linspace(-10,10,100)

activations = {
    "linear": x,
    "relu": np.maximum(0,x),
    "sigmoid": 1/(1+np.exp(-x)),
    "tanh": np.tanh(x),
}

plt.figure(figsize = (8,6))

for func in user_input:
    if func in activations:
        plt.plot(x,activations[func],label = func.capitalize())
    else:
        print(f"'{func}' is not a recognized function. ")
        
plt.legend()
plt.title("Activation Functions")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```
